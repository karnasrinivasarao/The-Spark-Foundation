```python
#Importing all Necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading the Data

```python
store_data = pd.read_csv('SampleSuperstore.csv')
```

```python
store_data.head()
```

```python
store_data.shape
```

```python
store_data.drop(columns='Postal Code', inplace=True)
```

```python
#Removing the all Duplicate Values from data

store_data.duplicated()
```

```python
store_data.drop_duplicates(inplace=True)
```

```python
#After removing Duplicates data has been reduced

store_data.shape
```

# Data Visualization with Ship Mode : 

```python
#Grouping Ship Mode column with Profit 

ship_mode1 = store_data.groupby(['Ship Mode'], as_index=False)['Profit'].sum()

#plotting the bar plot Ship Mode vs Profit

plt.figure(figsize=(8,6))
sns.barplot(ship_mode1['Ship Mode'], ship_mode1['Profit'])
plt.show()
```

### From abow graph we can see that who has prefered Ship Mode as Standard Class has more Profit and Some day has less Profit.

```python
#Grouping the Data by Ship Mode column with Sales and Profit

ship_mode2 = store_data.groupby(['Ship Mode'])[['Sales', 'Profit']].sum()
ship_mode2
```

```python
#plotting the bar plot Ship Mode vs Sales and Profit

ship_mode2.plot(kind='bar', figsize=(10,6))
plt.show()
```

## From aboe graph we can see that Standard Class has both more Profit and Sales.

```python
#Grouping the Data by Ship Mode column with Sales and Profit and Discount

ship_mode5 = store_data.groupby(['Ship Mode'])['Sales', 'Discount', 'Profit'].sum()
```

```python
#plotting the pie chart 

ship_mode5.plot.pie(autopct='%1.1f%%', label=ship_mode5.index, subplots=True, figsize=(20,10))
plt.show()
```

## From abow pie chart we can see that Standard Class has more Sales, Profit and Discount. 

## If we try to give some more Discount for First Class and Second Class there may be chancess of getting more Profit and more Sales.
